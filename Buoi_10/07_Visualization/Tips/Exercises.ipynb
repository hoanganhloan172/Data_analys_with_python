### Step 1. Import the necessary libraries:

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
tips = pd.read_csv(url)
print(tips.head())
```

### Step 4. Delete the Unnamed 0 column

```python
# Remove the 'Unnamed: 0' column
tips.drop(columns=['Unnamed: 0'], inplace=True)

# Print the DataFrame to verify the deletion
print(tips.head())
```

### Step 5. Plot the total_bill column histogram

```python
import matplotlib.pyplot as plt

# Plot histogram of the 'total_bill' column
plt.hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of total_bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
import matplotlib.pyplot as plt

# Create scatter plot of total_bill vs tip
plt.figure(figsize=(8, 6))
plt.scatter(tips['total_bill'], tips['tip'], color='orange', edgecolor='black', alpha=0.7)
plt.title('Scatter plot of total_bill vs tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
import matplotlib.pyplot as plt

# Create scatter plot with 'total_bill', 'tip', and 'size'
plt.figure(figsize=(10, 8))
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10, alpha=0.7, c='skyblue', edgecolors='black')
plt.title('Relationship between Total Bill, Tip, and Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(['Size'], loc='upper left')
plt.grid(True)
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set(style="whitegrid")

# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Relationship between Days and Total Bill')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create scatter plot with day as y-axis and tip as x-axis, differentiated by sex
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.title('Scatter Plot of Tip vs Day, Differentiated by Sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.legend(title='Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set(style="whitegrid")

# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Total Bill per Day, Differentiated by Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
import matplotlib.pyplot as plt

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Filter data for Dinner and Lunch
dinner_tips = tips[tips['time'] == 'Dinner']
lunch_tips = tips[tips['time'] == 'Lunch']

# Plot histogram for Dinner tips
axes[0].hist(dinner_tips['tip'], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title('Histogram of Tips for Dinner')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')

# Plot histogram for Lunch tips
axes[1].hist(lunch_tips['tip'], bins=10, color='salmon', edgecolor='black')
axes[1].set_title('Histogram of Tips for Lunch')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create two separate scatter plots for Male and Female
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Subplot for Male
sns.scatterplot(data=tips[(tips['sex'] == 'Male')], x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs Tip for Male')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Smoker')

plt.subplot(1, 2, 2)  # Subplot for Female
sns.scatterplot(data=tips[(tips['sex'] == 'Female')], x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs Tip for Female')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Smoker')

plt.tight_layout()
plt.show()
```
