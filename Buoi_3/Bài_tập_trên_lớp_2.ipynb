```python
#bai 6:
import random
def create_rainfall_dictionary():
    rainfall_dict = {}
    for month in range(1, 13):

        rainfall_values = [random.uniform(100, 4000) for _ in range(20)]

        rainfall_dict[month] = rainfall_values
    return rainfall_dict

rainfall_data = create_rainfall_dictionary()


for month, values in rainfall_data.items():

    print(f"Tháng {month}: {values}")
```

```python
#Bai7:

def create_dictionary(A, B):

    C = {}

    # Xử lý từ điển A

    for key, value in A.items():

        C[key] = value

    # Xử lý từ điển B

    for key, value in B.items():

        if key in C:

            # Nếu key đã xuất hiện trong A và B, gán value là max của A và B

            C[key] = max(C[key], value)

        else:

            # Nếu key chỉ xuất hiện trong B, gán value là value của B

            C[key] = value

    return C

# Ví dụ sử dụng

A = {'a': 10, 'b': 20, 'c': 30}

B = {'b': 25, 'c': 35, 'd': 40}

C = create_dictionary(A, B)
```
