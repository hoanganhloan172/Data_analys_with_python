### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Create a data dictionary that looks like the DataFrame below

```python
d = {'Name':pd.Series(['Bulbasaur','Charmander','Squirtle','Caterpie','pikachu']),
   'hp':pd.Series([255,126,25,293,999]),
   'evolution':pd.Series(["Ivysaur","Charmeleon","Wartortle","Metapod","raichu"]),
    'pokedex':pd.Series(["yes","no","yes","no","yes"]),
     "type": pd.Series(["grass","fire","water","bug","electric"])
}
```

### Step 3. Assign it to a variable called pokemon

```python
pokemon = pd.DataFrame(d)
pokemon.head()
```

### Step 4. Ops...it seems the DataFrame columns are in alphabetical order. Place  the order of the columns as name, type, hp, evolution, pokedex

```python
pokemon = pokemon[["Name","type","hp","evolution","pokedex"]]
pokemon.head()
```
